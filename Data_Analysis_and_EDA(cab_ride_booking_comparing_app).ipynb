```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=pd.read_excel('/content/drive/MyDrive/Colab Notebooks/cab customer data.xlsx')
```

**DATA ANALYSIS or Product Market Fit or Market Research**

The mobile users are increasing very rapidly in India so it is obvious that users of online cab riding app will also increase.

```python
plt.rcParams['figure.figsize']=(15,6)
```

```python
# plotting a graph to show how many people ever booked a cab or not

s=x['have you ever booked an online cab'].value_counts()
s.plot(kind='pie',explode=[0,0.5],autopct='%1.1f%%')

plt.title('ever booked a cab online')
```

This pie chart shows out of 100 people in random 79% has booked a cab online and 21% has not booked a cab online.

```python
# plotting a graph to show which is the prefered cab company

x['cab company usually prefer'].value_counts().plot(kind='bar')
plt.title('prefered company')
```

This bar chart give insights that ola is the first preference of people and second is uber and other local company cab services are at third. The margin between the three is not very much that shows that people do use cab services in India in large amount.

```python
# preference of vehicle type to travel

x['type of vehicle you prefer to travel'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('prefered vehicle type')
```

From this chart the insights we get that in India the people prefer cab services and then thier private vehicles and then public transport. So ultimately our users are in good amount and it is increasing as per statistics and facts on google and other sites.

```python
# plotting graph for usage of medium to book a cab service

x['booking method'].value_counts().plot(kind='bar')
plt.title('medium used')
```

As a result of the chart we can see clearly that the users are using app and website mostly to book a cab service and limited are booking a taxi physically or calling the company.

```python
# plotting graph to show how frequently people book cab services

x['travel period with cab'].value_counts().plot(kind='barh')
plt.title('frequency of using cab services')
```

We can say that in India the daily user and every week user of cab services count is significant by concluding this graph.

```python
# plotting a graph that shows the reason for using cab services

s=list(x['reason to travel with cab'].value_counts())
l=list(x['reason to travel with cab'].unique())
colors=['lightcoral', 'lightskyblue']
plt.pie(s, labels=l,colors=colors, autopct='%1.1f%%')
plt.title('reason to use cab service')
```

This pie chart shows that people uses the cab services mostly for travel for work and office travel.

```python
# plotting a graph to know that exactly how many percent users would like a app from which they can compare rides and book the respective rides

x['do you like a app from which you can compare rides'].value_counts().plot(kind='pie', explode=[0,0.3,0.4], autopct='%1.1f%%')
plt.title('User Preferences for our Application')
```

A vast majority of users will prefer a app or website that can compare different cab services company and give recomendations and book a cab also respectively.

**Reading Second Dataset**

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/cab data.csv')
df.head()
```

**Exploratory Data Analysis(EDA)**

```python
df.info()
```

```python
df.describe()
```

```python
# checking distribution, skewness, mean and median and mode condition for distance field


n=['distance', 'toll_amount']

for col in n:
  plt.figure()
  sns.distplot(df[col])
  plt.show()
```

```python
# checking outliers


for col in n:
  plt.figure(figsize=(8, 6))
  plt.boxplot(df[col])
  plt.xlabel(col)
  plt.show()
```

```python
# plotting number of passenger in a single ride


df['num_passengers'].value_counts().plot(kind='bar')
plt.title('passenger count')
```

We can see that above graphs and insights showed that user uses the cab services mostly for work related travel and this graph also shows that solo passenger trips are in great quantity than others.

```python
df['payment_method'].value_counts()
```

```python
# plotting payment methods to show the distribution

#payment methods are symbolized as---- 1 = Credit Card, 2 = Cash, 3 = Free ride, 4 = Disputed, 5 = Unknown, 6 = Void trip

df['payment_method'].value_counts().plot(kind='pie')
plt.title('payment medium')
```

```python
# plotting types of rides to show the distribution

# type of rides are symbolized as---- 1 = Standard, 2 = Airport, 3 = Connaught Place, 4 = Noida, 5 = Negotiated Fare, 6 = Pooled ride
print(df['rate_code'].value_counts())
df['rate_code'].value_counts().plot(kind='pie')
plt.title('ride_type')
```

The standard rides are booked in large amount and then airport rides are booked.
